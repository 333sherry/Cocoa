```python
def cuisinerec(people):
  profiles = {}
  individual = {}
  for person in people: #grabbing info from the database table
    query = f'SELECT * FROM {person}Cuisine'
    profiles[person] = pd.read_sql(query, con=engine)
    individual[person] = profiles[person].nlargest(3, 'Rating')['Cuisine'].tolist()
  combined = pd.concat([profiles[person].set_index('Cuisine')['Rating'] for person in people], axis=1)
  combined.columns = people
  combined['Mean'] = combined.mean(axis=1)
  combo = combined['Mean'].nlargest(3).index.tolist()

  row = [{'User': user, 'Top Cuisines': cuisines} for user, cuisines in individual.items()]
  indtable = pd.DataFrame(row)

  return combo, indtable
```

```python
def restrictioncheck(people):
  reslist = []
  for person in people:
    query = f'SELECT * FROM {person}Restrictions'
    profile = pd.read_sql(query, con=engine)

    for i in range(5):

      if int(profile['preference'][i]) == 1:
        reslist.append(profile['name'][i])

  return reslist
```

```python
def getcoord(results, geocode_key):
    # Function to get coordinates for an address
    def get_coordinates(address):
        # URL for the Geocoding API request
        url = 'https://maps.googleapis.com/maps/api/geocode/json'

        # Parameters for the API request
        params = {
            'address': address,
            'key': geocode_key
        }

        # Send GET request to the Geocoding API
        response = requests.get(url, params=params)
        data = response.json()

        # Extract latitude and longitude from the response
        if data['status'] == 'OK':
            location = data['results'][0]['geometry']['location']
            return (location['lat'], location['lng'])
        else:
            return None

    # Apply the function to each address in the DataFrame
    results['coordinates'] = results['location'].apply(get_coordinates)

    return results
```

```python
def counttags(df):
  all = []
  for cell in df['category'].dropna():
    s = [s.strip() for s in cell.split(',')]
    all.extend(s)
  count = pd.Series(all).value_counts()
  table = count.reset_index()
  table.columns = ['Tag', 'Count']
  return table
```

```python
def countprice(df):
  all = []
  for cell in df['price_point'].dropna():
    s = [s.strip() for s in cell.split(',')]
    all.extend(s)
  count = pd.Series(all).value_counts()
  table = count.reset_index()
  table.columns = ['Price Point', 'Count']
  return table
```
