```python
import pandas as pd

# Selecting all entries from the database
query = "SELECT * FROM food_drink_final"
df = pd.read_sql(query, con=engine)

# Function to extract state abbreviation from the address
def extract_state(address):
    try:
        # Assuming the state is the second to last element after splitting by ','
        # and that it is the last part when split by space
        parts = address.split(',')
        state_part = parts[-2].strip() if len(parts) > 1 else parts[-1].strip()
        return state_part.split()[-1]
    except IndexError:
        # Returning None or some default if the expected format is not met
        return None

# Applying the function to create a new column for state
df['state'] = df['location'].apply(extract_state)

# List of states to exclude
states_to_exclude = ["NJ", "CT", "PA"]

# Filtering out entries from the specified states
df = df[~df['state'].isin(states_to_exclude)]
df
```

```python
# Lets do this again because you never know...
remove_types = [
    'Strip Clubs', 'Zoos', 'Tennis', 'Swimming Pools', 'Museums',
    'Not specified', 'Performing Arts', 'Festivals', 'Gift Shops', 'Cinema',
    'Gyms', 'Cycling Classes', 'Arcades', 'Virtual Reality Centers',
    'Day Camps', 'Parks', 'Acupuncture', 'Alternative Medicine'
]

# Create a mask to filter rows
mask = df['category'].apply(lambda x: all(item not in x for item in remove_types))

# Special handling for 'Venues & Event Spaces' if it is the only entry
mask &= df['category'].apply(lambda x: 'Venues & Event Spaces' not in x or x.strip() != 'Venues & Event Spaces')

# Apply the mask
filtered_df = df[mask]

filtered_df
```

# Analysis on final restaurant data set
Now that the data is cleaned lets take a look at some possible trends

```python
query = "SELECT * FROM food_drink_final"
df = pd.read_sql(query, con=engine)
df
```

```python
# Lets look at what the average rating is based on borough and plot it
average_rating_by_borough = df.groupby('borough')['rating'].mean()
print(average_rating_by_borough)

import matplotlib.pyplot as plt
average_rating_by_borough.plot(kind='bar', color='skyblue', figsize=(10, 6))
plt.title('Average Restaurant Ratings by Borough')
plt.xlabel('Borough')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)

plt.show()
```

```python
ratings_by_borough = df.groupby('borough')['rating'].apply(list)

plt.boxplot(ratings_by_borough.values)
plt.xlabel('Borough')
plt.ylabel('Rating')
plt.title('Box Plot of Restaurant Ratings by Borough')
plt.xticks(range(1, len(ratings_by_borough) + 1), ratings_by_borough.index, rotation=45)
plt.grid(True)
plt.show()
```

```python
borough_counts = df['borough'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(borough_counts, labels=borough_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Restaurants by Borough')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

Manhattan has the most restaurants with the highest ratings. Brooklyn is a close second and third is Queens. The Bronx and Staten Island seem to have more consistent ratings with few outliers compared to Brooklyn, Manhattan, and Queens. Most restaurants lised are in Manhattan as seen in the pie char.

Lets see what the highest rated categories are...

```python
# Lets check to see what the top 10 highest rated catergories are
highest_rated_categories = df.groupby('category')['rating'].mean().sort_values(ascending=False).head(10)
print(highest_rated_categories)

highest_rated_categories.plot(kind='bar', color='purple', figsize=(10, 6))
plt.title('Average Restaurant Ratings by Category')
plt.xlabel('Category')
plt.ylabel('Average Rating')
plt.xticks(rotation=90)

plt.tight_layout() # to make sure spacing is good
plt.show()
```

Wow! There seems to be a lot of restaurants with an average rating of 5.0. Wonder what the distribution of ratings looks like...

```python
# Plotting the distribution of rating scores
import numpy as np

counts, bins, _ = plt.hist(df['rating'], bins=30, color='green', edgecolor='black')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('Distribution of Rating Scores')
plt.grid(True)

average_rating = df['rating'].mean()
median_rating = df['rating'].median()

plt.axvline(x=average_rating, color='red', linestyle='dashed', linewidth=1, label=f'Average Rating: {average_rating:.2f}')
plt.axvline(x=median_rating, color='blue', linestyle='dashed', linewidth=1, label=f'Median Rating: {median_rating:.2f}')

# Find the bin with the highest frequency
max_bin_index = np.argmax(counts)
highest_frequency = counts[max_bin_index]
bin_center = (bins[max_bin_index] + bins[max_bin_index + 1]) / 2
plt.scatter(bin_center, highest_frequency, color='red', label=f'Highest Frequency: {highest_frequency}', zorder=5)# adding marker

plt.legend()
plt.tight_layout() # to make sure spacing is good
plt.show()
```

```python
rating_counts = df['rating'].value_counts()

rating_percentage = (rating_counts / rating_counts.sum()) * 100
mask = rating_percentage < 3
other_count = rating_counts[mask].sum()
rating_counts = rating_counts[~mask]
rating_counts['Other'] = other_count

# Pie chart
labels = rating_counts.index
plt.figure(figsize=(8, 8))
plt.pie(rating_counts, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Ratings')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

```python
category_counts = df['category'].value_counts()

category_percentage = (category_counts / category_counts.sum()) * 100
mask = category_percentage < .9
other_count = category_counts[mask].sum()
category_counts = category_counts[~mask]
category_counts['Other'] = other_count

# Pie chart
labels = category_counts.index
plt.figure(figsize=(8, 8))
plt.pie(category_counts, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Categories')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

```python
plt.boxplot(df['rating'])
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('Box Plot of Rating Scores')
plt.grid(True)
plt.show()
```

Most ratings lie between 4.0 and 4.167 with a frequency of 2572. The average rating is 3.83 and median is 4.0. We can also see in the pie charts that most ratings are a 4.0, making up 8.2% of the total rating scores. Rating under what its seems a 2.4 are rare and considered outliers as show in the boxplot.

Lets see if the price point effects rating...

```python
# Assign numerical values to price point categories
price_mapping = {'$': 1, '$$': 2, '$$$': 3, '$$$$': 4, 'None': 0}  # Assuming 'None' for restaurants with no price point listed

# Map price point categories to numerical values
df['price_numeric'] = df['price_point'].map(price_mapping)

plt.figure(figsize=(10, 6))
plt.scatter(df['rating'], df['price_numeric'], alpha=0.5)
plt.title('Scatter Plot of Rating versus Price')
plt.xlabel('Rating')
plt.ylabel('Price Point')
plt.grid(True)
plt.show()
```

Seems that a restaurant is more likely to get a higher rating if their price point is higher. That said the rating distribution for restaurants with the lowest price point ($) is considerably wide. Having a lower price point does not mean you score lower.
